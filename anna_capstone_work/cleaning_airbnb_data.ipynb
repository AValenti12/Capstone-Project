```python
#getting rid of columns with at least half non NaN values
df = df.dropna(1,thresh=25000)
df.head(5)
```

```python
# Drop the null rows
df = df.dropna()
df.head()
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
# convert columns to numerical
df['host_response_rate'] = df['host_response_rate'].str.replace('%', '').astype('float') / 100
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str. replace(',', '').astype('float') / 100
df['security_deposit'] = df['security_deposit'].str.replace('$', '')
df['security_deposit'] = df['security_deposit'].str. replace(',', '').astype('float') / 100
df['cleaning_fee'] = df['cleaning_fee'].str.replace('$', '')
df['cleaning_fee'] = df['cleaning_fee'].str. replace(',', '').astype('float') / 100
df['extra_people'] = df['extra_people'].str.replace('$', '')
df['extra_people'] = df['extra_people'].str.replace(',', '').astype('float') / 100
```

```python
amenities = []
for i in df['amenities']:
    amenities.append(i.count(",") + 1)
df['amenities'] = amenities
```

```python
x = {'flexible':0,
 'moderate':1,
 'strict_14_with_grace_period':2,
 'super_strict_30':3,
 'super_strict_60':4,'strict':5,'t':1,'f':0}    
df = df.replace(x)
df['cancellation_policy'] = df['cancellation_policy'].astype('float') / 100
```

```python
is_success = []
for i in df['review_scores_value']:
    is_success.append(int(i > 5.))
df['is_success'] = is_success
```
